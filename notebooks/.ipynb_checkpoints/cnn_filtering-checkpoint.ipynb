#### Hits Occupancy Maps 

```python
plt.figure(figsize=(12,9))

binning = 200

theRange = ((-60,60),(-0.1,18.0))

n = np.histogram2d(data_sample["inZ"],data_sample["inR"],bins=binning,range=theRange)

cnorm=colors.LogNorm(1e-1,n[2].max()*10)

plt.hist2d(data_sample["inZ"],data_sample["inR"],bins=binning,cmap="viridis",norm=cnorm,range=theRange)
ax = plt.gca()
plt.title("Inner Hit Occupancy Map",fontsize=18,fontweight="bold")
plt.ylabel("Z (cm)",fontsize=16,fontstyle="italic")
plt.xlabel("R (cm)",fontsize=16,fontstyle="italic")
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)

plt.colorbar()

plt.figure(figsize=(12,9))

n = np.histogram2d(data_sample["outZ"],data_sample["outR"],bins=binning,range=theRange)

cnorm=colors.LogNorm(1e-1,n[2].max()*10)

plt.hist2d(data_sample["outZ"],data_sample["outR"],bins=binning,cmap="viridis",norm=cnorm,range=theRange)
ax = plt.gca()
plt.title("Outer Hit Occupancy Map",fontsize=18,fontweight="bold")
plt.ylabel("Z (cm)",fontsize=16,fontstyle="italic")
plt.xlabel("R (cm)",fontsize=16,fontstyle="italic")
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)

plt.colorbar()
```
